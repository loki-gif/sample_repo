```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("car_crashes")

print(df.head())
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=df.index, y="total", data=df)
plt.xticks(rotation=90)
plt.title("Total Car Crashes by State")
plt.show()
```

```python
sns.histplot(df["alcohol"], kde=True)
plt.title("Distribution of Alcohol-related Accidents")
plt.show()
```

```python
sns.scatterplot(x="speeding", y="total", data=df)
plt.title("Speeding vs Total Crashes")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

Overall Insight

The exploratory data analysis of the Car Crashes dataset revealed significant differences in road safety across U.S. states. The visualizations showed that fatal crash rates vary considerably, with some states experiencing much higher accident frequencies than others. Factors such as speeding and alcohol involvement were found to have a positive relationship with the total number of crashes, indicating that they are major contributors to road accidents. The analysis also highlighted variations in insurance premiums and losses, suggesting that states with higher accident rates often incur greater insurance costs. Correlation analysis helped identify relationships between different variables, while distributions and box plots revealed the presence of outliers and differences among states. Overall, the EDA provided valuable insights into accident patterns and risk factors, demonstrating how data visualization can support informed decision-making for improving road safety and traffic management.
